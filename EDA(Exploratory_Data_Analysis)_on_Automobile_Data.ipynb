```python
import pandas as pd
import numpy as np
data=pd.read_csv('https://raw.githubusercontent.com/nunnarilabs/ml/master/Automobile_Data/automobile_data.csv')
```

```python
data.replace('?',np.nan,inplace=True)
data.head(5)
```

```python
avg=data['normalized-losses'].astype('float').mean(axis=0)
print('Average of normalized-losses',avg)
```

```python
data['normalized-losses'].replace(np.nan,avg,inplace=True)
data
```

```python
avg_bore=data['bore'].astype(float).mean(axis=0)
data['bore'].replace(np.nan,avg_bore,inplace=True)
print(avg_bore)
data

#do the same for all the other null values
```

```python
data['num-of-doors'].value_counts()
```

```python
data.dropna(subset=['price'],axis=0,inplace=True)
data.reset_index(drop=True,inplace=True)
```

```python
avg_stroke=data['stroke'].astype(float).mean(axis=0)
data['stroke'].replace(np.nan,avg_stroke,inplace=True)

avg_hp=data['horsepower'].astype(float).mean(axis=0)
data['horsepower'].replace(np.nan,avg_hp,inplace=True)

avg_rpm=data['peak-rpm'].astype(float).mean(axis=0)
data['peak-rpm'].replace(np.nan,avg_rpm,inplace=True)
```

```python
data[['bore','stroke']]=data[['bore','stroke']].astype(float)
data[['normalized-losses']]=data[['normalized-losses']].astype(int)
data[['price']]=data[['price']].astype(float)
data[['peak-rpm']]=data[['peak-rpm']].astype(float)
data[['horsepower']]=data[['horsepower']].astype(float)
```

```python
avg_nod=data['num-of-doors'].mode()
data['num-of-doors'].replace(np.nan,avg_nod[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
from scipy import stats
hp_zscore=np.abs(stats.zscore(data['normalized-losses']))
hp_zscore
```

```python
threshold=hp_zscore.max()
threshold
```

```python
index_val=np.where(hp_zscore>=threshold)
data.iloc[index_val]['normalized-losses']
```

```python
new_data=data.drop(index=data.iloc[index_val]['normalized-losses'].index)
new_data
```
